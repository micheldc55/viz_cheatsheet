```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
means = [
    25.0, 38.2, 89.1, 93.3, 46.9, 62.1, 77.1, 
    33.1, 41.7, 85.6, 90.9, 51.2, 63.1, 85.4
]

color = 7 * ['blue'] + 7 * ['red']
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday'] * 2

df = pd.DataFrame({'mean': means, 'color': color, 'day_of_week': days})

grid = sns.catplot(data=df, row='color', x='day_of_week', y='mean', kind='bar', height = 4, aspect = 2)

for idx in range(len(df['color'].unique())):
    for container in grid.axes[idx, 0].containers:
        grid.axes[idx, 0].bar_label(container, fmt='%.1f%%')
```
